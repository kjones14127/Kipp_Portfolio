```python
thc=pd.read_csv('/Users/test/Documents/NonProfit/test_rename_stack2.csv', encoding="Latin-1")  

cols_to_split = ['full_time_only_salary','gender','job_title','minimium_education','minimium_years_experience','race','status']
cols_no_split = ['Unique_ID']
expanded_rows = {}

# need the number of splits to fill 
n_splits = len(thc[cols_to_split[0]].loc[0].split(";"))

# expand elements of columns with single values to fit the same format of the others 
for col in cols_no_split:
  thc[col] = thc[col].apply(lambda x: ''.join([str(x)+';' for i in range(n_splits)]))
  thc[col] = thc[col].apply(lambda x: x[:-1])

# iterate through every column to expand them 
for col in (cols_no_split + cols_to_split):

  expanded_rows[col] = thc[col].str.split(';').apply(pd.Series,1).stack().values

#print(expanded_rows)
df_final = pd.DataFrame.from_dict(expanded_rows,orient='index')
df_final.transpose()
 
```
