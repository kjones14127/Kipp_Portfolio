```python
import pandas as pd
table_csv_path = r"C:\Users\kippk\Documents\DataScience101_Docs\Project2_Data_Files\Project2_Export.csv"
df=pd.read_csv(table_csv_path, dtype={'FIPS':str})
pd.set_option('max_columns', 40)
pd.set_option("min_rows", 10)
df 
```

```python
# delete rows where birthMonth,birthDay, birthYear have null values
df=df.dropna(axis=0, subset=['birthMonth'])
df=df.dropna(axis=0, subset=['birthDay'])
df=df.dropna(axis=0, subset=['birthYear'])
df
```

```python
df['birthMonth'] = df['birthMonth'].astype(int)   #convert to date fields to integer
df['birthDay'] = df['birthDay'].astype(int)
df['birthYear'] = df['birthYear'].astype(int)
df.dtypes
#df
```

```python
#Concatenate DOB
cols=["birthYear","birthMonth","birthDay"]
df['DOB'] = df[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns") 
df 
```

```python
#Convert DOB data type to datetime
df['DOB'] = pd.to_datetime(df['DOB']) 
df.dtypes

```

```python
#Double check Dataframe
df
```

```python
#calculate Age column
now = pd.Timestamp('now')
df['age']='age'#
df['age'] = (now - df['DOB']).astype('<m8[Y]') 
df['age']=df['age'].astype(int)
```

```python
#Double check Dataframe
df
```

```python
#Drop Unnecessary columns
df=df.drop(['nameFirst','nameLast','birthMonth','birthDay','birthYear','playerID.1','deathYear','deathMonth','deathCountry',
            'deathDay','deathState','deathCity','retroID','bbrefID'], axis = 1)
df
```

```python
#Drop the rows where at least one element is missing.
#df=df.dropna()
# delete rows where birthCountry	birthState	birthCity have null values
df=df.dropna(axis=0, subset=['birthCountry'])
df=df.dropna(axis=0, subset=['birthState'])
df=df.dropna(axis=0, subset=['birthCity'])
```

```python
#Export to csv
df.to_csv(r'C:\Users\kippk\Documents\DataScience101_Docs\Project2_Data_Files\to_csv\baseball_stats_1.csv', index = None, header=True)
```
